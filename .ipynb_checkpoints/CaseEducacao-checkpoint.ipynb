Vamos tentar entender a influência dos antecedentes dos pais, preparação para teste etc. no desempenho dos alunos.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
# Ler base de dados
base_dados = pd.read_csv('estudantesperf.csv')

# Dimensão
base_dados.shape
```

```python
# Head
base_dados.head()
```

```python
# Ver campos nulos
nulos = base_dados.isnull()
plt.figure(figsize=(16,5))
plt.title('Analise campos nulos')
sns.heatmap(nulos, cbar=False);
```

```python
# ver/contar campos nulos
nulos.sum()
```
