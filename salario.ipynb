```python
# Libs Necessárias

# Libs para Modelagem e Matrizez
import numpy as np
import pandas as pd

# Libs para anaálises gráficas
import matplotlib.pyplot as plt
import seaborn as sns

# Lib para ignorar avisos
import warnings

# Desabilitando avisos
warnings.filterwarnings('ignore')
```

```python
# Ler dados
base_dados = pd.read_csv('Salary_Data.csv')
 # Verificando 
base_dados.head()
```

```python
# Renomear colunas
base_dados.rename( columns={
    'YearsExperience' : 'Xp',
    'Salary' : 'Renda'
}, inplace=True)
```

```python
# dimensão
base_dados.shape
```

```python
# ver campos nulos
base_dados.isnull().sum()
```

```python
# grafico de nulos
sns.heatmap(base_dados.isnull(), cbar=False);
```

```python
base_dados.describe()
```

```python
plt.figure(figsize=(10,6))
plt.title('Analise de Renda')
sns.kdeplot(base_dados['Renda'], color='green');
```

```python
plt.figure(figsize=(10,6))
plt.title('Analise da Xp')
sns.kdeplot(base_dados['Xp'], color='red');
```

```python
# Distribuição
plt.figure(figsize=(10,6))
sns.boxplot(base_dados['Renda']);
```

```python
# Distribuição
plt.figure(figsize=(10,6))
sns.boxplot(base_dados['Xp']);
```

```python
# Distribuição
sns.scatterplot(data=base_dados, x='Renda', y='Xp');
```

```python
sns.regplot(data=base_dados, x='Renda', y='Xp');
```

```python
# correlação
correlacao = base_dados.corr('pearson')
plt.figure(figsize=(5,5))
sns.heatmap(correlacao, annot=True, cbar=False);
```

```python
# correlação
correlacao = base_dados.corr('spearman')
plt.figure(figsize=(5,5))
sns.heatmap(correlacao, annot=True, cbar=False);
```
