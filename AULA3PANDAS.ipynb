```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
import pandas as pd
```

```python
df = pd.read_csv("meu_dataset_modificado_copia.csv")
```

```python
# cria a figura
plt.figure(figsize = (3,3))

#cria o plot
plt.plot([15 ,20, 99], [13,90,10],
         color = 'green', # mudar cor da linha do grafico
         linewidth = 2, #muda espessura da linha do graf
         linestyle = '--', #muda a linha pra tracejado
         label = 'BRAT' #renomear linha
         )

#inserir outra linha
plt.plot([1,50,100], [10,56,99],
         color = 'red',
         linewidth = 2,
         linestyle = '--',
         label = 'THE TORTURED POETS DEPARTAMENT' # legendar linha
         )
# labels - rótulos
# rótulo do eixo x
plt.xlabel('Dias')
# rótulo do eixo y
plt.ylabel('Reproduções')
# adicionar título
plt.title('Reproduções Spotify')
# grid lines (colocar a gradezinha)
plt.grid(alpha=0.3) #alpha muda a opacidade da linha
plt.legend(loc='lower left', bbox_to_anchor=(1.0, -0.15)) # legenda das linhas
plt.show()

```

```python
#plotar varios graficos
fig, ax =plt.subplots(2,2, figsize = (6,6))
# inserir linha na 1ra caixinha [0,0]
ax[0,0].plot([1,2,3], [5,5,9])
ax[0,0].set(xlabel = 'Eixo X', ylabel = 'Eixo Y')
# inserir linha na 2da caixinha [0,1]
ax[0,1].plot([1,2,3], [5,5,9])
# inserir linha na 3ra caixinha [1,0]
ax[1,0].plot([1,2,3], [5,5,9])
#inserir linha na 4ta caixinha
ax[1,1].plot([1,2,3], [5,5,9])
plt.tight_layout(pad=4)  # separa as caixinhas
```

```python
fig, ax = plt.subplots(1,2, figsize = (12,6))
sns.countplot(data = df,
              y= 'TP_ESTADO_CIVIL', ax = ax[0])
sns.countplot(data = df,
              y = 'TP_ESTADO_CIVIL', hue = 'TP_SEXO', ax = ax[1])
ax[0].set(ylabel = 'Estado Civil', xlabel= 'Quandidade', title= 'Estado Civil')
ax[1].set(ylabel = 'Estado Civil', xlabel= 'Quantidade', title= 'Estado Civil por Gênero')
ax[1].legend(title = 'Gênero')
fig.suptitle('Gênero')
fig.suptitle('Estado Civil dos Inscritos')
fig.tight_layout(pad=4);
```
